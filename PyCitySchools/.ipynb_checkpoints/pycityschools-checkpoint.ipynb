```python
#initializing code
import pandas as pd
schooldata=pd.read_csv("schools_complete.csv")
studentdata=pd.read_csv("students_complete.csv")

#district summary
totalschools=len(schooldata["School ID"])
totalstudents=len(studentdata["Student ID"])
totalbudget=schooldata["budget"].sum()/1000 #convert to thousands so that it doesn't display in log
averagemath=studentdata["math_score"].mean()
averagereading=studentdata["reading_score"].mean()

#initialize variables to be calculated
percentmath=0
percentreading=0
for item in studentdata["math_score"]:
    if item > 69:
        percentmath = percentmath+1
for item in studentdata["reading_score"]:
    if item > 69:
        percentreading = percentreading+1

percentmath=percentmath/totalstudents*100
percentreading=percentreading/totalstudents*100
percentoverall=(percentmath+percentreading)/2

districtsummary=pd.DataFrame({"Metric":['Total Schools','Total Students','Total Budget (thousands)','Average Math Score','Average Reading Score','% Passing Math','% Passing Reading','Overall Passing Rate'],
                              "Value":[totalschools,totalstudents,totalbudget,averagemath,averagereading,percentmath,percentreading,percentoverall]})
districtsummary
```

```python
#school summary
schoolsummary=schooldata[['school_name','type','size','budget']]
perstudentbudget=[]
schoolmathavg=[]
schoolreadingavg=[]
schoolmathpass=[]
schoolreadingpass=[]
schooloverallpass=[]

#populate our blank lists by doing 1 loop for each school
for i in range(totalschools):
    studentbudget=int(schoolsummary.iloc[i,3])/int(schoolsummary.iloc[i,2])
    perstudentbudget.append(studentbudget)
    schoolmath=[]
    schoolreading=[]
    mathpass=[]
    readingpass=[]
    #within each school loop, run through a student loop
    for x in range(len(studentdata)):
        #within the student loop, find the student that matches the school and grab their scores
        if studentdata.iloc[x,4] == schoolsummary.iloc[i,0]:
            schoolmath.append(studentdata.iloc[x,6])
            if studentdata.iloc[x,6] > 69:
                mathpass.append(studentdata.iloc[x,6])
            schoolreading.append(studentdata.iloc[x,5])
            if studentdata.iloc[x,5] > 69:
                readingpass.append(studentdata.iloc[x,5])
    #process those matching scores into the metrics we want on the list
    schoolmathavg.append(sum(schoolmath)/len(schoolmath))
    schoolreadingavg.append(sum(schoolreading)/len(schoolreading))
    schoolmathpass.append(len(mathpass)/len(schoolmath)*100)
    schoolreadingpass.append(len(readingpass)/len(schoolreading)*100)
    schooloverallpass.append((schoolmathpass[i]+schoolreadingpass[i])/2)
    
#append the finished lists back to the shool summary as new columns
schoolsummary['Per Student Budget']=perstudentbudget
schoolsummary['Average Math Score']=schoolmathavg
schoolsummary['Average Reading Score']=schoolreadingavg
schoolsummary['% Passing Math']=schoolmathpass
schoolsummary['% Passing Reading']=schoolreadingpass
schoolsummary['Overall Passing Rate']=schooloverallpass
schoolsummary
```

```python
#top performing schools
schoolsummary=schoolsummary.sort_values(by='Overall Passing Rate',ascending=False)
topschools=schoolsummary.iloc[0:5,:]
topschools
```

```python
#bottom performing schools
schoolsummary=schoolsummary.sort_values(by='Overall Passing Rate',ascending=True)
bottomschools=schoolsummary.iloc[0:5,:]
bottomschools
```

```python
#math scores by grade
mathgradesummary=schooldata[['school_name','type']]
gradesmath9th=[]
gradesmath10th=[]
gradesmath11th=[]
gradesmath12th=[]

for x in range(len(mathgradesummary)):
    school9th=[]
    school10th=[]
    school11th=[]
    school12th=[]
    for i in range(len(studentdata)):
        if studentdata.iloc[i,4] == mathgradesummary.iloc[x,0] and studentdata.iloc[i,3]=='9th': 
            school9th.append(studentdata.iloc[i,6])
        elif studentdata.iloc[i,4] == mathgradesummary.iloc[x,0] and studentdata.iloc[i,3]=='10th': 
            school10th.append(studentdata.iloc[i,6])
        elif studentdata.iloc[i,4] == mathgradesummary.iloc[x,0] and studentdata.iloc[i,3]=='11th': 
            school11th.append(studentdata.iloc[i,6])
        elif studentdata.iloc[i,4] == mathgradesummary.iloc[x,0] and studentdata.iloc[i,3]=='12th': 
            school12th.append(studentdata.iloc[i,6])
    gradesmath9th.append(sum(school9th)/len(school9th))
    gradesmath10th.append(sum(school10th)/len(school10th))
    gradesmath11th.append(sum(school11th)/len(school11th))
    gradesmath12th.append(sum(school12th)/len(school12th))

mathgradesummary['9th']=gradesmath9th
mathgradesummary['10th']=gradesmath10th
mathgradesummary['11th']=gradesmath11th
mathgradesummary['12th']=gradesmath12th
mathgradesummary
```

```python
#reading scores by grade
readinggradesummary=schooldata[['school_name','type']]
gradesreading9th=[]
gradesreading10th=[]
gradesreading11th=[]
gradesreading12th=[]

for school in readinggradesummary['school_name']:
    school9th=[]
    school10th=[]
    school11th=[]
    school12th=[]
    for i in range(len(studentdata)):
        if studentdata.iloc[i,4] == school and studentdata.iloc[i,3]=='9th': 
            school9th.append(studentdata.iloc[i,5])
        elif studentdata.iloc[i,4] == school and studentdata.iloc[i,3]=='10th': 
            school10th.append(studentdata.iloc[i,5])
        elif studentdata.iloc[i,4] == school and studentdata.iloc[i,3]=='11th': 
            school11th.append(studentdata.iloc[i,5])
        elif studentdata.iloc[i,4] == school and studentdata.iloc[i,3]=='12th': 
            school12th.append(studentdata.iloc[i,5])
    gradesreading9th.append(sum(school9th)/len(school9th))
    gradesreading10th.append(sum(school10th)/len(school10th))
    gradesreading11th.append(sum(school11th)/len(school11th))
    gradesreading12th.append(sum(school12th)/len(school12th))

readinggradesummary['9th']=gradesreading9th
readinggradesummary['10th']=gradesreading10th
readinggradesummary['11th']=gradesreading11th
readinggradesummary['12th']=gradesreading12th
readinggradesummary
```

```python
### Scores by School Spending
spending_bins = [0, 585, 615, 645, 675]
spendingschoolsummary = pd.DataFrame({'Spending ranges per student':["<$585", "$585-615", "$615-645", "$645-675"]})
col1=[]
col2=[]
col3=[]
col4=[]
col5=[]

for i in range(len(spending_bins)-1):
    subcol1=[]
    subcol2=[]
    subcol3=[]
    subcol4=[]
    subcol5=[]
    for x in range(len(schoolsummary['Per Student Budget'])):
        if schoolsummary['Per Student Budget'][x] >= spending_bins[i] and schoolsummary['Per Student Budget'][x] <spending_bins[i+1]: 
            subcol1.append(schoolsummary.iloc[x,5])
            subcol2.append(schoolsummary.iloc[x,6])
            subcol3.append(schoolsummary.iloc[x,7])
            subcol4.append(schoolsummary.iloc[x,8])
            subcol5.append(schoolsummary.iloc[x,9])
    col1.append(sum(subcol1)/len(subcol1))
    col2.append(sum(subcol2)/len(subcol2)) 
    col3.append(sum(subcol3)/len(subcol3))
    col4.append(sum(subcol4)/len(subcol4))
    col5.append(sum(subcol5)/len(subcol5))
    
spendingschoolsummary['Average math score']=col1
spendingschoolsummary['Average reading score']=col2
spendingschoolsummary['% Passing math']=col3
spendingschoolsummary['% Passing reading']=col4
spendingschoolsummary['% Overall passing rate']=col5
spendingschoolsummary
```
