```python
#initializing code
import pandas as pd
schooldata=pd.read_csv("schools_complete.csv")
studentdata=pd.read_csv("students_complete.csv")

#district summary
totalschools=len(schooldata["School ID"])
totalstudents=len(studentdata["Student ID"])
totalbudget=schooldata["budget"].sum()/1000 #convert to thousands so that it doesn't display in log
averagemath=studentdata["math_score"].mean()
averagereading=studentdata["reading_score"].mean()

#initialize variables to be calculated
percentmath=0
percentreading=0
for item in studentdata["math_score"]:
    if item > 69:
        percentmath = percentmath+1
for item in studentdata["reading_score"]:
    if item > 69:
        percentreading = percentreading+1

percentmath=percentmath/totalstudents*100
percentreading=percentreading/totalstudents*100
percentoverall=(percentmath+percentreading)/2

districtsummary=pd.DataFrame({"Metric":['Total Schools','Total Students','Total Budget (thousands)','Average Math Score','Average Reading Score','% Passing Math','% Passing Reading','Overall Passing Rate'],
                              "Value":[totalschools,totalstudents,totalbudget,averagemath,averagereading,percentmath,percentreading,percentoverall]})

#school summary
schoolsummary=schooldata[['school_name','type','size','budget']]
perstudentbudget=[]
schoolmathavg=[]
schoolreadingavg=[]
schoolmathpass=[]
schoolreadingpass=[]
schooloverallpass=[]

#populate our blank lists by doing 1 loop for each school
for i in range(totalschools):
    studentbudget=int(schoolsummary.iloc[i,3])/int(schoolsummary.iloc[i,2])
    perstudentbudget.append(studentbudget)
    schoolmath=[]
    schoolreading=[]
    mathpass=[]
    readingpass=[]
    #within each school loop, run through a student loop
    for x in range(len(studentdata)):
        #within the student loop, find the student that matches the school and grab their scores
        if studentdata.iloc[x,4] == schoolsummary.iloc[i,0]:
            schoolmath.append(studentdata.iloc[x,6])
            if studentdata.iloc[x,6] > 69:
                mathpass.append(studentdata.iloc[x,6])
            schoolreading.append(studentdata.iloc[x,5])
            if studentdata.iloc[x,5] > 69:
                readingpass.append(studentdata.iloc[x,5])
    #process those matching scores into the metrics we want
    schoolmathavg.append(sum(schoolmath)/len(schoolmath))
    schoolreadingavg.append(sum(schoolreading)/len(schoolreading))
    schoolmathpass.append(len(mathpass)/len(schoolmath)*100)
    schoolreadingpass.append(len(readingpass)/len(schoolreading)*100)
    schooloverallpass.append((schoolmathpass[i]+schoolreadingpass[i])/2)
    
#append the finished lists back to the shool summary
schoolsummary['Per Student Budget']=perstudentbudget
schoolsummary['Average Math Score']=schoolmathavg
schoolsummary['Average Reading Score']=schoolreadingavg
schoolsummary['% Passing Math']=schoolmathpass
schoolsummary['% Passing Reading']=schoolreadingpass
schoolsummary['Overall Passing Rate']=schooloverallpass
schoolsummary

#top performing schools
```
